```python
#!/usr/bin/python

import zipfile

def docx_replace(old_file,new_file,rep):
    zin = zipfile.ZipFile (old_file, 'r')
    zout = zipfile.ZipFile (new_file, 'w')
    for item in zin.infolist():
        buffer = zin.read(item.filename)
        if (item.filename == 'word/document.xml'):
            res = buffer.decode("utf-8")
            for i in range(4):
                for r in rep[i]:
                    res = res.replace(r,rep[i][r])
                    #print(rep[i][r], sep=' ')
            buffer = res.encode("utf-8")
        zout.writestr(item, buffer)
    zout.close()
    zin.close()

import os, shutil

def clear_folder(folder):
    for the_file in os.listdir(folder):
        file_path = os.path.join(folder, the_file)
        try:
            if os.path.isfile(file_path):
                os.unlink(file_path)
            #elif os.path.isdir(file_path): shutil.rmtree(file_path)
        except Exception as e:
            print(e)

input_file = open("/Users/ruslan/Documents/Военка/prog/data/all_students.txt", "r")
info_list = input_file.read().split('\n')
N = len(info_list)

clear_folder('/Users/ruslan/Documents/Военка/prog/Numberly_cards/cards_corrected')

FIO = []
surname = []
name = []
middle_name = []
nationality = []
birth_place = []
education = []
specialization = []
sports_category = []
marital_status = []
address = []
mobile = []
height = []
cap_size = []
mask_size = []
uniform_size = []
shoes_size = []
mom_name = ["0"]*N
mom_adress = ["0"]*N
mom_phone = ["0"]*N
zvan = []
sold_num = []



chemists = {"Безменов", "Гевондян", "Гутенев", "Демидов", "Ефремов", "Магоммедов", "Маклаков", "Махмутов", "Овезов", 
            "Рыжкин", "Станишевский", "Таранченко", "Фролов", "Венгрус"}

to_correct = {"Шипилов", "Шаров", "Шамсутдинов", "Трушников", 
              "Токтаев", "Сайфуллин", "Кувакин", "Комягин", 
              "Комиссаров", "Карпов", "Исаков", "Заверткин", 
              "Еремиев", "Елкин", "Гузенков", "Гитолендия", 
              "Волков", "Бахтин", "Балагазов", "Багурцев", 
              "Арсеньев", "Абизгильдин", "Чижов", "Соболев", 
              "Отрыванов", "Каримов", "Камалов", "Галкин",
              "Бухарский", "Безменов", "Андрейцев", "Белоусов"}

cnt = 0
for i in range(N):
    cur_info = info_list[i].split('\t')
    cur_FIO = cur_info[0].split()
    if cur_FIO[0] in to_correct:
        surname.append(cur_FIO[0])
        name.append(cur_FIO[1])
        if cur_FIO[2] == "Аббас" or cur_FIO[2] == "Ариф":
            middle_name.append(cur_FIO[2]+" "+"оглы")
        else:
            middle_name.append(cur_FIO[2])
        #print(i, cur_FIO)
        nationality.append(cur_info[1])
        birth_place.append(cur_info[2])
        #education.append(cur_info[3])
        if surname[i-cnt] in chemists or surname[i-cnt] == "Огородников" and name[i-cnt] == "Роман":
            education.append("2018")
        else:
            education.append("2016")
        specialization.append(cur_info[4])
        if cur_info[5] == "нет":
            sports_category.append("")
        else:
            sports_category.append(cur_info[5])
        marital_status.append(cur_info[6])
        address.append(cur_info[7])
        mobile.append(cur_info[8])
        height.append(cur_info[9])
        cap_size.append(cur_info[10])
        mask_size.append(cur_info[11])
        uniform_size.append(cur_info[12])
        shoes_size.append(cur_info[13])
        if len(cur_info) >= 15:
            mom_name[i-cnt] = cur_info[14] + ","
        if len(cur_info) >= 16:
            mom_adress[i-cnt] = cur_info[15]+ ","
        if len(cur_info) >= 17:
            mom_phone[i-cnt] = cur_info[16]
    else:
        cnt+=1
    
FIO_file = open("/Users/ruslan/Documents/Военка/prog/data/fio_~final.txt", "r")
FIO_database_list = FIO_file.read().split('\n')
vus_file = open("/Users/ruslan/Documents/Военка/prog/data/VUS-numer.txt", "r")
vus_database_list = vus_file.read().split('\n')
bday_file = open("/Users/ruslan/Documents/Военка/prog/data/BD_new.txt", "r")
bday_database_list = bday_file.read().split('\n')
FIO_badge_file = open("/Users/ruslan/Documents/Военка/prog/data/badges.txt", "r")
FIO_badge = FIO_badge_file.read().split('\n')
FIO_badge_dict = dict()

for x in FIO_badge:
    FIO_badge_dict[x.split('\t')[0].strip()] = x.split('\t')[1]
#print(FIO_vus_dict)

for x in FIO_badge:
    if FIO_badge_dict[x.split('\t')[0].strip()] == "":
        FIO_badge_dict[x.split('\t')[0].strip()] = "____________"
        
# print(FIO_badge_dict)
    
FIO_vus_dict = dict()
FIO_bday_dict = dict()
FIO_rank_dict = dict()

for i in range(len(FIO_database_list)):
    FIO_vus_dict[FIO_database_list[i]] = vus_database_list[i]
    if vus_database_list[i] == "188 (182)":
        FIO_rank_dict[FIO_database_list[i]] = "сержант"
    else:
        FIO_rank_dict[FIO_database_list[i]] = "рядовой"
    FIO_bday_dict[FIO_database_list[i]] = bday_database_list[i]
    
#print(FIO_rank_dict)

N = len(to_correct)+1
for i in range(0, N - 4, 4):
    full_name = ["0"]*4
    stud_dict = ["0"]*4
    turbo_word = surname[i+0] + "_" + surname[i+1] + "_" + surname[i+2] + "_" + surname[i+3]
    print(turbo_word)
    for j in range(4):
        full_name[j] = surname[i+j] + " " + name[i+j] + " " + middle_name[i+j]
        stud_ind = str(j+1)
        #print("soldnum"+stud_ind, FIO_badge_dict[full_name[j]])
        stud_dict[j] = {"fam"+stud_ind:surname[i+j],
                     "imya"+stud_ind:name[i+j],
                     "otch"+stud_ind:middle_name[i+j],
                     "place"+stud_ind:birth_place[i+j],
                     "rank"+stud_ind:FIO_rank_dict[full_name[j]],
                     "zhet"+stud_ind:FIO_badge_dict[full_name[j]],
                     "birth"+stud_ind:FIO_bday_dict[full_name[j]],
                     "momim"+stud_ind:mom_name[i+j],
                     "mompl"+stud_ind:mom_adress[i+j],
                     "momtel"+stud_ind:mom_phone[i+j]
                    }
        #print(mom_name[i+j])
    #print(turbo_word)
    docx_replace("/Users/ruslan/Documents/Военка/prog/Numberly_cards/numberly_template.docx", 
                 "/Users/ruslan/Documents/Военка/prog/Numberly_cards/cards_corrected/"+turbo_word+".docx", 
                 stud_dict)
input_file.close()
FIO_file.close()
vus_file.close()
bday_file.close()
FIO_badge_file.close()

print("Success!")



```
